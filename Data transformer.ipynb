```python
import pandas as pd
```

```python
df = pd.read_csv("CSVs/SD 25 Dec to 30 Dec.csv")
word_to_remove = "DT"
df["Sender"] = df["Sender"].str.replace(f"{word_to_remove}", "", case=True, regex=False).str.strip()

df.head()
```

```python
df2 = df.copy()
df2.drop(["Time", "Zone"], axis="columns", inplace=True)
df2.head()
```

```python
df2["First_7_Words"] = df2["Message"].str.split().str[:7].str.join(" ")
df2.head()
```

```python
search_word = "SD"
df2["flag"] = df2["First_7_Words"].str.contains(fr"\b{search_word}\b", case=False, regex=True)
df2.head()
```

```python
df3 = df2[df2["flag"] == True].drop(["Message", "First_7_Words"], axis="columns")
print(f"Count: {df3.shape[0]}")
df3
```

```python
df3.to_csv("SD Pitch.csv", header=False, index=False)
```
